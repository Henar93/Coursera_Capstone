```python
#Data set URL 
url = 'https://opendata-ajuntament.barcelona.cat/data/dataset/808daafa-d9ce-48c0-925a-fa5afdb1ed41/resource/4cc59b76-a977-40ac-8748-61217c8ff367/download/districtes_i_barris_170705.csv'
Barcelona_data = pd.read_csv(url, encoding='utf-8', sep=",")
Barcelona_data.columns = ['BoroughID','Borough','NeighborhoodID','Neighborhood']
Barcelona_data.head()
```

```python


# Function that gets geographic coordinates of each neighborhood of Barcelona
def get_Coord(row):
    
    location = None
    c=0
    # Sometimes geopy is not able to get the coordinates at first try.
    while location == None and c<100:
        try:
            location = geolocator.geocode(row['Neighborhood'] + ', ' + row['Borough'] + ', Barcelona')
        except:
            pass
        c+=1

    if location == None:
        return (np.NaN, np.NaN)
    else:
        return (location.latitude, location.longitude)
```

```python
# Apply get_Coord to each dataframe row
Barcelona_data[''] = Barcelona_data.apply(get_Coord, axis=1)
Barcelona_data['Latitude'] = Barcelona_data[''].apply(lambda x: x[0])
Barcelona_data['Longitude'] = Barcelona_data[''].apply(lambda x: x[1])
Barcelona_data.drop(columns="",inplace=True)
Barcelona_data.head()
```
